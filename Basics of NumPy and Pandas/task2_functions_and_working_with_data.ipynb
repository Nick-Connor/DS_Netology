# Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:  

оценка 2 и ниже — низкий рейтинг;  
оценка 4 и ниже — средний рейтинг;  
оценка 4.5 и 5 — высокий рейтинг.  
Результат классификации запишите в столбец class.  

```python
data_ratings = pd.read_csv ("ratings.csv")
data_ratings.head()
```

```python
def ratings_filter(row):
  rating = row["rating"]
  if rating <= 2:
    row["class"] = "Фильм с низким рейтингом"
  elif rating <= 4:
    row["class"] = ("Фильм со средним рейтингом")
  else:
   row["class"] = ("Фильм с высоким рейтингом")
  return row
```

```python
data_rated_classes = data_ratings.apply(ratings_filter, axis=1)
data_rated_classes.head()
```

# Задание 2
Используйте файл keywords.csv.  

Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.  

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:  

geo_data = {  

'Центр': ['москва', 'тула', 'ярославль'],  

'Северо-Запад': ['петербург', 'псков', 'мурманск'],  

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']  
}  

Результат классификации запишите в отдельный столбец region.  

```python
data_regions = pd.read_csv ("keywords.csv")
data_regions.head()
```

```python
keywords = pd.read_csv("keywords.csv")


geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}


def define_region(keyword):
    """
    Определяет регион по ключевому слову.
    Если город не найден, возвращает 'undefined'
    """
    keyword_lower = str(keyword).lower()

    for region, cities in geo_data.items():
        for city in cities:
            if city in keyword_lower:
                return region

    return 'undefined'


keywords['region'] = keywords['keyword'].apply(define_region)


print(keywords.head(10))
print(f"\nРаспределение по регионам:")
print(keywords['region'].value_counts())
```
