# Задание 1
Датасет для домашнего задания находится в материалах занятия - файлы для домашнего задания «Библиотека Pandas».  

Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
# Соединяем фильмы и оценки
data = ratings.merge(movies, on="movieId")

data.head()
```

```python
# Считаем количество оценок 5.0 по каждому фильму
result = data[data["rating"] == 5.0]["title"].value_counts().idxmax()

print("Фильм с наибольшим числом оценок 5.0:", result)
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
data_power = pd.read_csv('power.csv')
data_power.head()
```

```python
# Фильтрация данных
filtered_data = data_power[
    (data_power['country'].isin(['Lithuania', 'Latvia', 'Estonia'])) &
    (data_power['category'].isin([4, 12, 21])) &
    (data_power['year'].between(2005, 2010)) &
    (data_power['quantity'] > 0)
]

# Суммирование потребления
total_consumption = filtered_data['quantity'].sum()

print(f"Суммарное потребление стран Прибалтики: {total_consumption}")
```
