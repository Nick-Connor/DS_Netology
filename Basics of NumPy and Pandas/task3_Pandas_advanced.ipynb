# Задание 1
Для датафрейма log из материалов занятия создайте столбец source_type по правилам:  

если источник traffic_source равен Yandex или Google, то в source_type ставится organic;  
для источников paid и email из России ставим ad;  
для источников paid и email не из России ставим other;  
все остальные варианты берём из traffic_source без изменений.  

```python
df_log = pd.read_csv("visit_log.csv", sep = ";")
df_log.head()
```

```python
df_log['source_type'] = df_log['traffic_source']

# Yandex или Google → organic
df_log.loc[df_log['traffic_source'].isin(['yandex', 'Yandex', 'google', 'Google']), 'source_type'] = 'organic'

# paid или email из России → ad
df_log.loc[(df_log['traffic_source'].isin(['paid', 'email'])) & (df_log['region'] == 'Russia'), 'source_type'] = 'ad'

# paid или email не из России → other
df_log.loc[(df_log['traffic_source'].isin(['paid', 'email'])) & (df_log['region'] != 'Russia'), 'source_type'] = 'other'

df_log.head(20)
```

# Задание 3
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.  

```python
df_ratings = pd.read_csv("ratings.csv")
df_ratings.head(10)
```

```python
df_ratings['date'] = pd.to_datetime(df_ratings['timestamp'], unit='s')
df_ratings.head()
```

```python
# Группируем по пользователям и считаем количество оценок
user_stats = df_ratings.groupby('userId').agg({
    'timestamp': ['min', 'max', 'count']  # мин, макс время и количество оценок
})
user_stats.head()
```

```python
# Упрощаем структуру столбцов
user_stats.columns = ['min_timestamp', 'max_timestamp', 'rating_count']
# Вычисляем время жизни для каждого пользователя
user_stats['lifetime'] = user_stats['max_timestamp'] - user_stats['min_timestamp']

# Фильтруем пользователей с более чем 100 оценками
active_users = user_stats[user_stats['rating_count'] > 100]

# Считаем среднее время жизни
average_lifetime = active_users['lifetime'].mean()

print(f"Среднее время жизни пользователей с >100 оценок: {average_lifetime:.2f} секунд")
print(f"Количество таких пользователей: {len(active_users)}")
```

# Задание 4
Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе «Материалы для лекции “Продвинутый pandas”» ---- Ноутбуки к лекции «Продвинутый pandas»).

Нужно сформировать две таблицы:  

таблицу с тремя типами выручки для каждого client_id без указания адреса клиента;  
аналогичную таблицу по типам выручки с указанием адреса клиента.  

Дана статистика услуг перевозок клиентов компании по типам:  
  
rzd - железнодорожные перевозки  
auto - автомобильные перевозки  
air - воздушные перевозки  
client_base - адреса клиентов  

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# Таблица с тремя типами выручки без адреса клиента
revenue_without_address = (rzd
    .merge(auto, on='client_id', how='outer')
    .merge(air, on='client_id', how='outer')
    .fillna(0)  # NaN заменяется на 0 для клиентов без определенного типа выручки
)
revenue_without_address
```

```python
# Таблица с тремя типами выручки с указанием адреса клиента
revenue_with_address = (revenue_without_address
    .merge(client_base, on='client_id', how='left')
)
revenue_with_address
```

```python
# Переупорядочиваем столбцы для лучшей читаемости
revenue_with_address = (revenue_without_address
    .merge(client_base, on='client_id', how='left')
    [['client_id', 'address', 'rzd_revenue', 'auto_revenue', 'air_revenue']]
)
revenue_with_address
```
